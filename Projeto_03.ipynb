```python
#Importação das Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#Configurações Pandas
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Criando o DataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Visualizando as 10 primeiras linhas
df.head(10)
```

```python
#Analisando as quantidades de linhas e colunas
df.shape
```

```python
#Analisando os tipos de Dados
df.dtypes
```

```python
#Qual foi a Receita Total?
receita_total = round(df["Valor Venda"].sum(), 2)
receita_total
```

```python
#Qual o Custo Total?
df["Custo Pedido"] = df["Custo Unitário"]*df["Quantidade"]
custo_total = round(df["Custo Pedido"].sum(), 2)
custo_total
```

```python
#Qual o Lucro Total?
df["Lucro"] = df["Valor Venda"] - df["Custo Pedido"]
lucro_total = round(df["Lucro"].sum(), 2)
lucro_total
```

```python
#Criando uma coluna com o total de dias para enviar um produto
df["Tempo Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head()
```

```python
#Extraindo apenas os dias
df["Tempo Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
#Qual a média de envio por marca?
df.groupby("Marca")["Tempo Envio"].mean()
```

```python
#Há valores faltando?
df.isnull().sum()
```

```python
#Qual o lucro por Ano e Por Marca?
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de Produtos Vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#Gráfico do Lucro por Ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Total Vendas por Ano")
plt.xlabel("Ano")
plt.ylabel("Lucro");
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro por Mês no Ano de 2009")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro por Marca no Ano de 2009")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro por Classe no Ano de 2009")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
